# Model Forecasting dengan fbprophet

```python
from fbprophet import Prophet
train = pd.read_csv('https://raw.githubusercontent.com/annisaarss/Trade-in-goods/main/Data%20impor%20vietnam.csv')
train.info()
```

```python
train.head()
```

```python
# mengubah tipe data ds menjadi datetime
train['ds'] = pd.to_datetime(train['ds'], dayfirst=True)
train.info()
```

```python
# membuat model
model = Prophet(interval_width = 0.95)
model.fit(train)
```

```python
# membuat prediksi
future = model.make_future_dataframe(periods = 7, freq = 'Y')
pred = model.predict(future)

fig = model.plot(pred)
```

```python
pred[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

Diprediksi bahwa vietnam akan mengalami kenaikan impor, sehingga vietnam merupakan salah satu rekomendasi negara tujuan ekspor Indonesia
